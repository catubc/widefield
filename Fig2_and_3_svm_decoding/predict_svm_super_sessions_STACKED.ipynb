```python
##############################################
######### LOOP OVER ALL CONDITIONS ###########
##############################################

# this function is for making rewarded/unrewarded times and generating the final data+label .npz files
root_dir = '/media/cat/4TBSSD/yuki/'

#
animal_ids = ['IA1','IA2','IA3','IJ1','IJ2','AQ2']

#
n_sec_lockouts = [0,1,2,3,6,9,12,15,18,21,24,27,30]
#n_sec_lockouts = [2]

#
body_feats_lockouts = [[],[0],[1],[5],[0,1,5]]


# loop over the lockouts 
# 
for animal_id in animal_ids:
    
    # 
    for n_sec_lockout in n_sec_lockouts:

        #
        for body_feats_lockout in body_feats_lockouts:
            # skip 1,2,3 n-sec-lockouts and the empty lockouts
            if len(body_feats_lockout)==0:
                if n_sec_lockout in [1,2,3]:
                    print ("skipping redundatn computation")
                    print ("         animal id: ", animal_id, "  n-sec: ", n_sec_lockout, 
                            " body feats: ", body_feats_lockout)
                    continue

            print ('')
            print ("animal id: ", animal_id, "  n-sec: ", n_sec_lockout, 
                    " body feats: ", body_feats_lockout)
            
            
            
            ########################################
            ########## LOAD SESSIONS ###############
            ########################################
            
            # 
            fname = os.path.join(root_dir,
                                 animal_id,
                                 'super_sessions',
                                 'alldata_body_and_nonreward_lockout_'+str(n_sec_lockout)+'secLockout_'+
                                 str(body_feats_lockout)+'bodyfeats.npz')
            
            # 
            data = np.load(fname, allow_pickle=True)
            if data['trials'].shape[0]==0:
                continue
            stacks_all = np.vstack(data['trials']).transpose(0,2,1)
            stacks_all_random = np.vstack(data['random']).transpose(0,2,1)

            ##########################################
            ############# RUN SVM ####################
            ##########################################
            svm = PredictSVMChoiceSuperSession()
            svm.root_dir = root_dir
            #svm.random_flag = False  # shuffle data to show baseline
            svm.animal_id = animal_id

            #
            svm.imaging_rate = 30        # MUST SET THIS MANUALLY FOR NOW

            # svm parameters
            svm.xvalidation = 10  # KFold xvalidation step
            svm.data_split = 0.9  # split 80/20 data for training
            svm.method = 'sigmoid'  # method used for computing SVM

            # run-time parameters
            # window parameters
            svm.sliding_window = svm.imaging_rate      # number of frames in sliding window
            svm.gpu_flag = False
            svm.parallel = True
            svm.n_cores = svm.xvalidation
            #
            #features = [3,4]
        #     0 Retrosplenial area, dorsal part, layer 1 - right
        #     1 Retrosplenial area, lateral agranular part, layer 1 - right
        #     2 Primary visual area, layer 1 - right
        #     3 Primary somatosensory area, upper limb, layer 1 - right
        #     4 Primary somatosensory area, lower limb, layer 1 - right
        #     5 Primary somatosensory area, barrel field, layer 1 - right
        #     6 Secondary motor area, layer 1 - right
        #     7 Primary motor area, Layer 1 - right
        #     8 Primary motor area, Layer 1 - left
        #     9 Secondary motor area, layer 1 - left
        #     10 Primary somatosensory area, barrel field, layer 1 - left
        #     11 Primary somatosensory area, lower limb, layer 1 - left
        #     12 Primary somatosensory area, upper limb, layer 1 - left
        #     13 Primary visual area, layer 1 - left
        #     14 Retrosplenial area, lateral agranular part, layer 1 - left
        #     15 Retrosplenial area, dorsal part, layer 1 - left

            features = np.arange(16)
            #features = [0,1,3,4,6,7,8,9,11,12,14,15]  # exclude visual and barrel cortex

            start_ = 0
            end_ = 100000

            t1 = 0
            t2 = 930

            # 
            stacks_all = stacks_all[start_:end_,features, t1:t2]
            stacks_all_random = stacks_all_random[start_:end_,features, t1:t2]

            #
            #print (stacks_all.shape, 
            #       stacks_all_random.shape)

            # 
            # fname_out = fname
            svm.predict3(stacks_all,
                         stacks_all_random,
                         fname)       

    
```

```python


#classification of time (10-class)
X_l=data_leverpress #  [n_trials, n_areas, n_time_steps]
X_l=X_l.transpose((0,2,1))


X_R_l= X_l.reshape(-1,X_l.shape[1]*X_l.shape[2])
normal_X_l = preprocessing.normalize(X_R_l)
n_X_l=normal_X_l.reshape(X_l.shape[0],X_l.shape[1],X_l.shape[2])
X_l=n_X_l

X_l=X_l[:,:300,:]
X_k=X_l[:,0:30,:]
X_k=X_k.reshape(X_k.shape[0],X_k.shape[1]*X_k.shape[2])
for i in range(30,271,30):
    X_t = X_l[:,i:i+30,:]
    X_t=X_t.reshape(X_t.shape[0],X_t.shape[1]*X_t.shape[2])
    X_k=np.concatenate((X_k,X_t),axis=0)

y_ct=np.zeros(data_leverpress.shape[0])
for i in range(1,10):
    Z_ct=i*np.ones(data_leverpress.shape[0])
    y_ct=np.concatenate((y_ct,Z_ct))

# 
X_tSVM=X_k # 10s * 30frams
y_tSVM=y_ct # labels

#10-fold confusion matrix
clf = svm.SVC() # Non-linear classifier
ten_svm=[]
conf_matrix_ten=[]
kf = KFold(n_splits=10,random_state=None, shuffle=True) 
for train_index, test_index in kf.split(X_tSVM):
    X_train_k, X_test_k = X_tSVM[train_index], X_tSVM[test_index] 
    y_train_k, y_test_k = y_tSVM[train_index], y_tSVM[test_index]
    clf.fit(X_train_k, y_train_k)
    score=clf.score(X_test_k, y_test_k)
    y_predicted=clf.predict(X_test_k)
    cm=confusion_matrix(y_test_k,y_predicted)
    confusion_m=cm.T # make each row be the prediction
    conf_matrix_norm = confusion_m.astype('float') / confusion_m.sum(axis=1)[:,np.newaxis] #calculate the precision
    conf_matrix_norm = np.nan_to_num(conf_matrix_norm)
    ten_svm.append(score)
    conf_matrix_ten.append(conf_matrix_norm)
np.save('conf_10_'+str(name)+'_'+str(jobid)+'.npy',conf_matrix_ten)
sc=np.mean(conf_matrix_ten,axis=0) # mean of confusion matrix
np.save('conf_avg_'+str(name)+'_'+str(jobid)+'.npy',sc)
```
